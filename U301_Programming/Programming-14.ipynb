## Bubble Sort Pseudocode Using For-For Loop

```{code}
:linenos:
PROCEDURE bubble_sort(items)

    // Initialise the variables
    num_items = LEN(items)
  
    // Pass through the array of items n-1 times
    FOR pass_num = 1 TO num_items - 1
        // Perform a pass
    	FOR index = 0 TO num_items - 2  
       	    // Compare items to check if they are out of order
            IF (items[index] > items[index + 1]) THEN
            	// Swap the items
            	temp = items[index]
            	items[index] = items[index + 1]
            	items[index + 1] = temp 
            ENDIF
      	NEXT index
    NEXT pass_num
ENDPROCEDURE 
```

## Bubble Sort Python Code Using For-For Loop

````python
# Version A
numbers = [6, 5, 3, 1, 8, 7, 2, 4]
last_index = len(numbers) - 1
comparison_count = 0

print("Initial state:", numbers)

for i in range(last_index):
    for j in range(last_index):
        comparison_count += 1
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
    print(f"After pass {i + 1}: ", numbers)

print("Total comparisons in Version A:", comparison_count)
````

````python
# Version B
numbers = [6, 5, 3, 1, 8, 7, 2, 4]
last_index = len(numbers) - 1
comparison_count = 0

print("Initial state:", numbers)

for i in range(last_index):
    for j in range(last_index - i):
        comparison_count += 1
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
    print(f"After pass {i + 1}: ", numbers)

print("Total comparisons in Version B:", comparison_count)
````

## Bubble Sort Pseudocode Using While-For Loop

```{code}
:linenos:
PROCEDURE bubble_sort_version_2(items)

    // Initialise the variables
    num_items = LEN(items)
    swapped = True
  
    // Repeat while one or more swaps have been made
    WHILE swapped == True
        swapped = False
        // Perform a pass
    	FOR index = 0 TO num_items - 2
       	    // Compare items to check if they are out of order
            IF items[index] > items[index + 1] THEN
            	// Swap the items
            	temp = items[index]
            	items[index] = items[index + 1]
            	items[index + 1] = temp
                swapped = True
            ENDIF
	NEXT index
    ENDWHILE
ENDPROCEDURE
```

:::{tip} Exercise 4
Compare the While-For Bubble Sort to For-For Bubble Sort:

- How many loops it has
- What is the main difference in the While-loop?
- What data structure do we use here?
- How do we know the number of items in this structure?
- In which lines, do we swtich values?
:::

````python
def bubble_sort_version_2(items):

    # Initialise the variables
    num_items = len(items)
    swapped = True

    # Repeat while one or more swaps have been made
    while swapped == True:
        swapped = False
        # Perform a pass
        for index in range(0, num_items - 1):
            # Compare items to check if they are out of order
            if items[index] > items[index + 1]:
                # Swap the items
                temp = items[index]
                items[index] = items[index + 1]
                items[index + 1] = temp
                swapped = True


numbers = [6, 5, 3, 1, 8, 7, 2, 4]
bubble_sort_version_2(numbers)
print(numbers)
````

## Pseudocode for Linear Search

Figure 2.26 

```
ALGORITHM linearSearch()
{ Purpose: Searches through a list of element }
{ Output: Boolean, True if item found, False if not }
BEGIN
    Input: searchList, searchItem
    found <- FALSE
    FOR eachItem in searchList DO
        IF eachItem = searchItem THEN
            found <- TRUE
            BREAK  { exit loop once found }
        ENDIF
    ENDFOR
    RETURN found
END
```

## Python for Linear Search

````python
def linearSearch(searchList, searchItem):
    """
    Searches through a list of elements for a specified item.

    :param searchList: List of elements to search through.
    :param searchItem: Item to search for in the list.
    :return: Boolean, True if item found, False if not.
    """
    found = False
    for eachItem in searchList:
        if eachItem == searchItem:
            found = True
            break  # Exit loop once found
    return found

# Example usage:
searchList = [12, 8, 31, 1, 77, 75, 18]
searchItem = 77
result = linearSearch(searchList, searchItem)
print("Item found:", result)
````
